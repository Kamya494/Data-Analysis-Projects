## Statistical Analysis

```python
print('mean',np.mean(close_price))
print('median',np.median(close_price))
print('min',np.min(close_price))
print('max',np.max(close_price))
print('std',np.std(close_price))
print('var',np.var(close_price))
print('mode',st.mode(close_price))
```

```python
print('mean',np.mean(open_price))
print('median',np.median(open_price))
print('min',np.min(open_price))
print('max',np.max(open_price))
print('std',np.std(open_price))
print('var',np.var(open_price))
print('mode',st.mode(open_price))
```

```python
print(np.percentile(close_price,25),
np.percentile(open_price,75),
np.percentile(open_price,90),
np.percentile(open_price,50))
```
