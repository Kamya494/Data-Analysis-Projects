## Boolean Filtering

```python
#Students whose Math marks are above 80.
maths=arr[:,0]
```

```python
maths[maths>80]
```

```python
print(np.mean(arr))
```

```python
#All Math scores above 80.

arr[:,0][(arr[:,0]>80)]
```

```python
#Students who scored below 40 in at least one subject.

arr[np.any(arr<40,axis=1)]
```

```python
#Students who scored above 90 in any subject.

arr[np.any(arr>90,axis=1)]
```

```python
#Students who scored at least 60 in every subject.

arr[np.all(arr>=60,axis=1)]
```

```python
#All students whose average marks are above 75.

arr[np.mean(arr,axis=1)>75]
```

```python
#Students whose marks are greater than 60 in Maths but less than 80 in English.

arr[(arr[:,0]>60)&(arr[:,3]<80)]
```

```python
#If a student's average marks are greater than 70 then he/she passes, else,fail.

data['Pass/Fail']=np.where(np.mean(arr,axis=1)>70,'Pass','Fail')
```
