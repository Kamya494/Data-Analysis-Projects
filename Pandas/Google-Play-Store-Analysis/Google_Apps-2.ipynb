## Exploratory Data Analysis

```python
data[data['Android Ver'].isna()]
```

```python
data['Android Ver']=data['Android Ver'].fillna('Unknown')
```

```python
print(data['Android Ver'].isna().sum())
```

```python
data[data['Type'].isna()]
```

```python
data[data['Rating'].isna()]
```

```python
data['Installs']=data['Installs'].str.strip('+')
```

```python
data['Installs']
```

```python
data['Installs']=data['Installs'].str.replace(',','')
```

```python
data['Installs']
```

#### Data in row 10472 was found having mis-aligned column values and the category and genre could not be found, making the retrieval of actual data in this row difficult hence it is deleted from the dataset.

```python
data=data.drop(10472)
```

```python
data['Installs']=data['Installs'].astype(int)
```

```python
data['Installs'].dtype
```

```python
data['Type'].value_counts()
```

```python
print(data['Rating'].max())
```

#### 10 highest rated apps.

```python
data.sort_values('Rating',ascending=False).head(10)
```

```python
grp_installs=data.groupby('Category')['Installs'].sum()
```

```python
grp_installs
```

#### Top 10 categories by install.

```python
top_ten=grp_installs.sort_values(ascending=False).head(10)
print(top_ten)

```

#### A bar plot representing the top 10 categories.

```python

plt.figure(figsize=[15,5])
plt.barh(top_ten.index,top_ten.values)
plt.xlabel('Category')
plt.ylabel('Installs')
plt.title('Top 10 Categories by Total Installs')
plt.show()
```

```python
data['Content Rating'].unique()
```

```python
data['Content Rating'].value_counts()
```

#### Total installs in each Content Rating category.

```python
grp_content=data.groupby('Content Rating')['Installs'].sum()
```

```python
grp_content
```

#### A bar plot representing total installs in each Content Rating category.

```python

plt.bar(grp_content.index,grp_content.values,edgecolor='black')
plt.xlabel('Content Rating')
plt.ylabel('Installs')
plt.xticks(rotation=45)
plt.show()
```

#### Average rating by category.

```python
grp_avg=data.groupby('Category')['Rating'].mean()
```

```python
grp_avg
```

#### A histogram showing the distribution of app ratings.

```python
plt.hist(data['Rating'],bins=20)
plt.show()
```

#### Number of apps in each category.

```python
app_category=data['Category'].value_counts()
```

```python
app_category
```

#### A bar plot representing the number of apps in each category.

```python
plt.figure(figsize=[20,8])
plt.barh(app_category.index,app_category.values)
plt.title('Number of Apps in Each Category')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.show()
```
