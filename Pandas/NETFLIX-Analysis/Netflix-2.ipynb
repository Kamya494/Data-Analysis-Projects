## Exploratory Data Analysis

#### No. of shows in each rating category.

```python
rating_cat=data['rating'].value_counts()
```

```python
rating_cat
```

```python
plt.barh(rating_cat.index,rating_cat.values)
plt.xlabel('Rating')
plt.ylabel('Content')
plt.title('No. of shows in each rating category')
plt.show()
```

#### Movie vs TV show

```python
type_content=data['type'].value_counts()
```

```python
type_content
```

```python
plt.pie(type_content.values,labels=type_content.index,autopct="%.2f")
plt.show()
```

#### Distribution of release year.

```python
plt.hist(data['release_year'],bins=20)
plt.show()
```

#### The histogram shows that a large proportion of Netflix titles were released around 2020.

#### Content added each year.

```python
data['year_added']=data['date_added'].dt.year
```

```python
data['year_added']
```

```python
year=data['year_added'].value_counts().sort_index()
```

```python
year
```

```python
plt.plot(year.index,year.values)
plt.xlabel('Year')
plt.ylabel('Number of content added')
plt.show()
```

#### Content additions peaked around 2019 before declining.
