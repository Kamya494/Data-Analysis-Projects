## Data Cleaning and Business Decisions

```python
data['Month']=pd.DatetimeIndex(data['InvoiceDate']).month
```

```python
data
```

```python
data['Sales']=data.Quantity	* data.UnitPrice
```

```python
data
```

```python
actual_sales=data[(data['StockCode']!='D')&(data['Description']!='Adjust bad debt')]
```

```python
actual_sales
```

#### Here we excluded all Discounts and Bad Debts.

```python
grp_month=actual_sales.groupby('Month')['Sales'].sum()
```

```python
grp_month
```

```python

print(grp_month.idxmax())
```

#### Net sales were highest during the month of November.
#### A possible explanation is the start of the holiday shopping season, with customers purchasing gifts ahead of Christmas and New Year.


```python
grp_country=actual_sales.groupby('Country')['Sales'].sum()
```

```python
grp_country
```

```python
grp_country.idxmax()
```

#### Highest net sales have been reported from United Kingdom.

```python
grp_country_max=actual_sales.groupby(['Country','Month']).agg({'Sales':'sum'})
```

```python
grp_country_max.head(20)
```

```python
grp_country_max.index
```

#### Data representing the Month is which each Country made their maximum sales.

```python
grp_country_max.groupby('Country')['Sales'].idxmax()
```

```python
actual_sales['Description'].unique()
```

```python
grp_product=actual_sales.groupby('Description').agg({'Sales':'sum'})
```

```python
grp_product.head(10)
```

```python
grp_product.sort_values('Sales',ascending=False).head(10)
```

#### During exploratory analysis, DOTCOM POSTAGE and POSTAGE were identified as a shipping charge rather than a retail product.

```python
actual_sales[actual_sales['Description']=='DOTCOM POSTAGE']
```

```python
actual_sales[actual_sales['Description']=='POSTAGE']
```

```python
actual_sales[actual_sales['Description']=='Manual']
```

#### Here Manual seems to be some manual adjustments made so it is excluded from the analysis of best-seller product.


```python
grp_product=grp_product.reset_index()
```

```python
best_seller=grp_product[(grp_product['Description']!='Manual')&(grp_product['Description']!='DOTCOM POSTAGE')&(grp_product['Description']!='POSTAGE')]
```

```python
best_seller
```

#### The top 10 best selling products.

```python
best_seller.sort_values('Sales',ascending=False).head(10)
```
